```python
import requests
import matplotlib.pyplot as plt

macbeth = [requests.get('http://www.gutenberg.org/cache/epub/2264/pg2264.txt').text]
#print(type(macbeth))
#print(len(macbeth))
#print(macbeth[:2000])


# Your code here
#for words in macbeth:
#word_split.append(words)

word_split = macbeth[0].split(" ")
    
#create dictionary
#iterate through dictionary
#update count key
word_count = set()

def get_wordcount(text):
    for word in word_split:
        if word != '':
            word_count.add((word, word_split.count(word)))
    return word_count

#Problem with the sorted values being added to the dictionary the same amount of times as the "Count" value....
word_counts = get_wordcount(word_split)

word_counts_sorted = sorted(word_counts,key=lambda x: x[1], reverse=True)
top_25 = word_counts_sorted[:25]

x_values = []
y_values = []

for word, count in top_25:
    x_values.append(word)
    y_values.append(count)


plt.figure(figsize=(25,10))
plt.bar(x_values, y_values)
plt.xlabel('Top 25 Words')
plt.ylabel('Count')
plt.title('Top 25 Word Counts in Macbeth')
plt.rcParams.update({'font.size': 12})
plt.show()

# Pseudo-code Outline
# Split the transcript into words
# Create a dictionary
# Iterate through the text of Macbeth
# Update word counts
# Create Bar Graph
# Include descriptive titles and labels
```
